```python
import pandas as pd #loads the relevant package
```

```python
#Load and change data types - long format
df = pd.read_csv("allmatchesv2.csv")
df['Actual'] = df['Actual'].str.replace(',', '').astype(float)
df['Budget'] = df['Budget'].str.replace(',', '').astype(float)
df.to_csv("Matchday_summary.csv", index=False)
```

```python
#Convert long format to wide format
budget_pivot = df.pivot_table(index='Match', columns=['Group', 'Metric'], values='Budget', aggfunc='sum')
budget_pivot.columns = [f"{group}_{metric}_Budget" for group, metric in budget_pivot.columns]
# Pivot actual column
actual_pivot = df.pivot_table(index='Match', columns=['Group', 'Metric'], values='Actual', aggfunc='sum')
actual_pivot.columns = [f"{group}_{metric}_Actual" for group, metric in actual_pivot.columns]

# Combine the pivoted DataFrames
result_df = pd.merge(budget_pivot.reset_index(), actual_pivot.reset_index(), on='Match')
group_metrics = [(group, metric) for group in df['Group'].unique() for metric in df['Metric'].unique()]
order = []

for group, metric in group_metrics:
    if f"{group}_{metric}_Budget" in result_df.columns:
        order.append(f"{group}_{metric}_Budget")
    if f"{group}_{metric}_Actual" in result_df.columns:
        order.append(f"{group}_{metric}_Actual")

# Rearrange columns based on custom order
result_df = result_df[['Match'] + order]
result_df
```

```python
result_df.to_csv("dawideway.csv", index=False)
```
