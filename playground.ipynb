```python
def generate_strings(input_list):
    output_list = []
    for s in input_list:
        output_list.append(s + " - Budget")
        output_list.append(s + " - Actual")
    return ", ".join(output_list)

input = ["FSE", "Group", "Single", "Total Paid", "Comp", "Total Distributed", "Drop Count", "Show Rate", "ATP", "Gross Ticket Revenue", "Amusement Taxes", "Park District Fee", "MLS Share of Revenue", "Net Ticket Revenue",
"Units Sold", "Avg. Per Unit", "Revenue Generated", "Less Projected Expenses", "Net Premium", "Merchandise (Net)", "Concessions (Net)", "Parking (Net)", "Total Net Per Cap (Net)", "Merchandise Per Cap (Net)", "Concessions Per Cap (Net)", "Parking Per Cap (Net)",
"Net Per Caps", "Merchandise (Gross)", "Concessions (Gross)", "Parking (Gross)", "Total Gross Per Cap", "Merchandise Per Cap (Gross)", "Concessions Per Cap (Gross)", "Parking Per Cap (Gross)", "Gross Per Caps", "Total Net MD Revenue"]
output = generate_strings(input)
print(output)
```

```python
vert = pd.read_csv("Matchday Summary Data Template.csv")
vert
```

```python
vert = vert[["Metric", "Budget", "Actual", "Match"]]
vert
```

```python
#long to wide

budget_pivot = vert.pivot_table(index='Match', columns='Metric', values='Budget', aggfunc='sum')
budget_pivot.columns = [f"{col} - Budget" for col in budget_pivot.columns]
# Pivot actual column
actual_pivot = vert.pivot_table(index='Match', columns='Metric', values='Actual', aggfunc='sum')
actual_pivot.columns = [f"{col} - Actual" for col in actual_pivot.columns]

# Combine the pivoted DataFrames
result_df = pd.concat([budget_pivot, actual_pivot], axis=1).reset_index()
result_df
```
